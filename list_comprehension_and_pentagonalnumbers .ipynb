```python
list1=[n*(3*n-1)//2 for n in range(1,101)]
list2=[]
list3=[]
def intersection(l1,l2):
    return [x for x in l1 if x in l2]
    
for i in list1:
    for j in list1:
        if (i+j) in list1:
            list2.append([i,j])
        if abs(i-j) in list1:
            list3.append([i,j])

print("list of addition \n",list2)
print("list of substraction \n",list3)
print("intersection is \n",intersection(list2,list3))
            
```
