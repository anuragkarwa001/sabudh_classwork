```python
def straight_flush(value,suit):
    listx=['2','3','4','5','6','7','8','9','T','J','Q','K','A']
    consec=[]
    for i in range(0,9):
        listp=[]
        for j in range(0,5):
            listp.append(listx[i+j])
        consec.append(set(listp))
    value=set(value)
    if value in consec and len(set(suit))==1:
        return True
    else:
        return False
    
def royal_flush(value,suit):
    suit=set(suit)
    if len(suit)==1:
        value=set(value)
        if value=={'T','J','Q','K','A'}:
            return True
    
    return False

def  flush(value,suit):
    suit=set(suit)
    if len(suit)==1:
        return True
    else:
        return False
def four_of_kind(value,suit):
    for i in value:
        if value.count(i)==4:
            return True 
    return False 

def straight(value,suit):
    listx=['2','3','4','5','6','7','8','9','T','J','Q','K','A']
    consec=[]
    for i in range(0,9):
        listp=[]
        for j in range(0,5):
            listp.append(listx[i+j])
        consec.append(set(listp))
    value=set(value)
    if value in consec:
        return True
    else:
        return False
    
    
def three_of_kind(value,suit):
    for i in value:
        if len(set(value))==3:
            if value.count(i)==3:
                return True
    return False

def full_house(value,suit):
    count=0 
    
    for i in value:
        if value.count(i)==3:
            count=3
            
    value=set(value)
    if len(value)==2 and count==3:
        return True
    else:
        return False

def one_pair(value,suit):
    for i in value:
        if value.count(i)==2 and len(set(value))==4:
            return True
    return False
def two_pair(value,suit):
    temp=set(value)
    for i in temp:
        if value.count(i)==2 :
            temp=temp-{i}
            for j in temp:
                if value.count(j)==2 and len(set(value))==3:
                    return True
                
    return False
def highest_card(value,suit):
    temp=[]
    for i in value:
        if i=='A':
            temp.append(14)
        elif i=='K':
            temp.append(13)
        elif i=='Q':
            temp.append(12)
        elif i=='J':
            temp.append(11)
        elif i=='T':
            temp.append(10)
        else:
            temp.append(i)
        
        return int(max(temp))
    
```
